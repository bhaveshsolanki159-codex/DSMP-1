```python
# find all the final winners
final = ipl['MatchNumber'] == 'Final'
new_df = ipl[final]
new_df[['Season','WinningTeam']]

ipl[ipl['MatchNumber']=='Final'][['Season','WinningTeam']] # shortcut
```

```python
# how many super over finishes have occured
ipl[ipl['SuperOver']=='Y']['SuperOver'].shape[0]
```

```python
# how many matches has csk won in kolkata
ipl[(ipl['City'] == 'Kolkata')&(ipl['WinningTeam']=='Chennai Super Kings')]['WinningTeam'].shape[0]
```

```python
# Toss winner is match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```
