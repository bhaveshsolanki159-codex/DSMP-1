```python
import streamlit as st
import pandas as pd

df = pd.read_csv('archive/startup_funding.csv')
df
```

```python
sorted(df['Startup Name'].unique().tolist()) # startup names
```

```python
df['Investors Name'] = df['Investors Name'].fillna('undisclosed') # assign missing value as undisclosed
sorted(df['Investors Name'].unique().tolist())
```

```python
df.info()
```

```python
# droping remarks col

df.drop(columns=['Remarks'], inplace=True)
df
```

```python
# seting index as sr no
df.set_index(keys='Sr No',inplace=True)
```

```python
# rename col
df.rename(columns={'Date dd/mm/yyyy':'date',
                   'Startup Name':'startup',
                   'Industry Vertical':'vertical',
                   'SubVertical':'subvertical',
                   'City  Location':'city',
                   'Investors Name':'investors',
                   'InvestmentnType':'round',
                   'Amount in USD':'amount'
                   }, inplace=True)
```

```python
# USD -> cR
df['amount'] = df['amount'].fillna('0')

df['amount'] = df['amount'].str.replace(',','')
df['amount'] = df['amount'].str.replace('undisclosed','0')
df['amount'] = df['amount'].str.replace('unknown','0')
df['amount'] = df['amount'].str.replace('Undisclosed','0')
df['amount'] = df['amount'].str.replace('+','')



```

```python
df = df[df['amount'].str.isdigit()]

df['amount'] = df['amount'].astype('float')
```

```python
def to_inr(dollor):
    inr = dollor * 82.5
    return inr/10000000

df.head()
```

```python
df['amount'] = df['amount'].apply(to_inr)
df.info()
```

```python
df['date'] = df['date'].str.replace('05/072018','05/07/2018')
```

```python
df['date'] = pd.to_datetime(df['date'],errors='coerce')
df['date'].dt.year
```

```python
df.head()
```

```python
df.info()
```

```python
# removing rows containing a missing values
df = df.dropna(subset=[ 'date','startup', 'vertical','city','investors','round','amount'])
df.info()
```
